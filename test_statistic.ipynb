```python
import pandas as pd
import numpy as np
from model.taiji.chiper import TaijiChiper
from tkinter.filedialog import askopenfilename
%matplotlib inline

filename = askopenfilename()

with open(filename, 'rb') as f:
    data = bytearray(f.read())
key = bytearray('kriptografitaiji'.encode())
tj_ecb = TaijiChiper(key, mode='ecb')
tj_cbc = TaijiChiper(key, mode='cbc')

ecb_chiper_data = tj_ecb.encrypt(data)
cbc_chiper_data = tj_cbc.encrypt(data)

df_ecb_chiper = pd.DataFrame(np.array(ecb_chiper_data))
df_cbc_chiper = pd.DataFrame(np.array(cbc_chiper_data))
df_plain = pd.DataFrame(np.array(data))

df_plain[0].hist()
```
